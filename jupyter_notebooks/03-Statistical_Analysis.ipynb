* 1.Descriptive Statistics

1.1 Principles of Statistics, Probability, and Data Analysis
Statistics is the branch of science used to collect, organize, analyze, and interpret data. The fundamental statistical concepts in the data analysis process are as follows:

1.1.1 Mean
The mean is obtained by dividing the sum of all values in a dataset by the number of elements.

Here, each data point is represented, and is the total number of elements in the dataset.

1.1.2 Median
The median is the middle value in an ordered dataset. If the number of elements is odd, it is the exact middle value; if even, it is the average of the two middle values.

1.1.3 Mode
The mode is the most frequently occurring value in a dataset. A dataset may have multiple modes or no mode at all.

1.1.4 Standard Deviation
Standard Deviation measures the amount of variation or dispersion of a dataset. A low standard deviation indicates that the data points are close to the mean, while a high standard deviation means the data points are spread out over a larger range.


1.1.5 Probability Distributions
Probability distributions represent the likelihood of an event occurring. One of the most common distributions is the normal distribution. The normal distribution is bell-shaped and represents the distribution of many natural phenomena.


Statistical Analysis with Python
Now, we will apply the concepts above by writing code in Jupyter Notebook.

Calculating Mean, Median, Standard Deviation, and Variance

```python
import pandas as pd
import numpy as np

#  Load the data
data = pd.read_csv(r"C:\Users\balla\OneDrive\Documents\CapStoneProject_2025\Data\Combined_Cleaned.csv")

# Select numerical columns
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns

# Compute basic statistics
mean_values = data[numeric_columns].mean()
median_values = data[numeric_columns].median()
std_values = data[numeric_columns].std()
var_values = data[numeric_columns].var()

print("üìå Mean Values:\n", mean_values)
print("\nüìå Median Values:\n", median_values)
print("\nüìå Standard Deviation Values:\n", std_values)
print("\nüìå Variance Values:\n", var_values)
```

* 1.3 Hypothesis Testing (T-Test)

Now, let's perform a t-test to determine whether there is a statistically significant difference between the means of two data groups.. One of the most commonly used tests is the t-test, which measures whether the difference in means between two groups is significant.It's appropriate when you want to test if there is a statistically significant difference between two independent or related groups (e.g., comparing the test scores of two classes).

The hypothesis testing process follows these steps:

Null Hypothesis (H‚ÇÄ): The assumption that there is no effect or no difference between groups.

Alternative Hypothesis (H‚ÇÅ): The assumption that there is a significant effect or difference between groups.

P-value: The probability of observing the data if the null hypothesis is true. If the p-value is less than the significance level (usually 0.05), the null hypothesis is rejected.


ANOVA (Analysis of Variance): Used to compare the means of three or more groups. It's suitable when you're testing if at least one of the group means differs significantly from others (e.g., comparing the test scores of three different classes).

Normal Distribution Graph
Let's plot a histogram to visualize the normal distribution of happiness scores.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import numpy as np

# Veriyi alma
happiness_scores = data["Happiness_score"].dropna()

# Histogram √ßizimi
plt.figure(figsize=(10, 5))
sns.histplot(happiness_scores, kde=True, bins=30, color="skyblue", stat="density")

# Normal daƒüƒ±lƒ±m eƒürisini ekleme
mean = happiness_scores.mean()
std = happiness_scores.std()
x = np.linspace(happiness_scores.min(), happiness_scores.max(), 100)
y = stats.norm.pdf(x, mean, std)
plt.plot(x, y, color="red", label="Normal Distribution")

# Grafik detaylarƒ±
plt.xlabel("Happiness Score")
plt.ylabel("Density")
plt.title("Normal Distribution of Happiness Scores")
plt.legend()
plt.show()

```
